```python
import joblib
import numpy as np
import pandas as pd

model = joblib.load("modelo_agua1.joblib")

def validar_input(nombre, minimo, maximo):
    while True:
        try:
            valor = float(input(f"Ingresa {nombre} ({minimo}-{maximo}): "))
            if minimo <= valor <= maximo:
                return valor
            else:
                print(f"Valor fuera de rango. Debe estar entre {minimo} y {maximo}")
        except ValueError:
            print("Ingresa un número válido")

print("=== Predicción de Potabilidad del Agua ===\n")

ph            = validar_input("pH",               0.0,   14.0)
hardness      = validar_input("Hardness",         0.0,  500.0)
solids        = validar_input("Solids",           0.0, 60000.0)
chloramines   = validar_input("Chloramines",      0.0,   15.0)
sulfate       = validar_input("Sulfate",          0.0,  500.0)
conductivity  = validar_input("Conductivity",     0.0,  800.0)
organic_carbon= validar_input("Organic_carbon",   0.0,   30.0)
trihalomethanes=validar_input("Trihalomethanes",  0.0,  130.0)
turbidity     = validar_input("Turbidity",        0.0,   10.0)

X = pd.DataFrame([{
    "ph": ph,
    "Hardness": hardness,
    "Solids": solids,
    "Chloramines": chloramines,
    "Sulfate": sulfate,
    "Conductivity": conductivity,
    "Organic_carbon": organic_carbon,
    "Trihalomethanes": trihalomethanes,
    "Turbidity": turbidity
}])

prediction = model.predict(X)[0]
resultado = "Potable" if prediction == 1 else "No potable"

print(f"\n=== Resultado ===")
print(f"El agua es: {resultado}")
```
